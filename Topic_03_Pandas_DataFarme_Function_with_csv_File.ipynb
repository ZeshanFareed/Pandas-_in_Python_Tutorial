# **DataFrame.aggregate**

```python
import pandas as pd
import numpy as np
data1 = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b': [6, 7, 8, 9, 10]})
print("Data1:")
print(data1)
print('Aggregate')
data2 = data1.agg('sum')
print(data2)
```

```python
import pandas as pd
import numpy as np
data1 = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b': [6, 7, 8, 9, 10]})
print("Data1:")
print(data1)
print('Aggregate')
data2 = data1.aggregate(['mean', 'median','max','min','sum','std','var','corrcoef','ptp'])
print(data2)
```

```python
import pandas as pd

# Existing DataFrame
df = pd.DataFrame({
    "NAME": ["IMRAN", "MUJHAID", "ALI", "SANWAL", "FAHEEM"],
    "ID": [2143, 4255, 4687, 5876, 7653],
    "AMOUNT": [5000, 3000, 5600, 34332, 6534],
    "PERCENTAGE": [95, 94, 80, 99, 70]
})

# New DataFrame to append
data2 = pd.DataFrame({
    'Country': ['Dubai', 'Dubai', 'Jampur'],
    'Age': [11, 12, 13],
    'Salary': [2300, 2400, 2500]
})

print("Concat")

# Correct way to concatenate DataFrames
data_combined = pd.concat([df, data2], ignore_index=True)
print(data_combined)
```
